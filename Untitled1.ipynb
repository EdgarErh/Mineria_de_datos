#####Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print (promedio)
print (promedio_ponderado)
```

#####Mediana

```python
print (mediana)
```

#####Varianza y Desviación Estándar

```python
df.head(7)
```

```python
# A NumPy
edad = df['edad'].to_numpy(dtype=float)
prog_py = df['prog_python'].to_numpy(dtype=bool)
usa_colab = df['usa_colab'].to_numpy(dtype=bool)
cuenta_gh = df['cuenta_github'].to_numpy(dtype=bool)
```

```python
print(f'ndarray edad primeros cinco elementos: {edad[:5]}')
print(f'ndarray prog_py primeros cinco elementos: {prog_py[:5]}')
print(f'ndarray usa_colab primeros cinco elementos: {usa_colab[:5]}')
print(f'ndarray cuenta_gh primeros cinco elementos: {cuenta_gh[-5:]}')
```

```python
# Descriptivos de edad (NumPy)
summary_age = {
    'n': int(edad.size),
    'min': float(np.min(edad)),
    'p25': float(np.percentile(edad, 25)),
    'mediana': float(np.median(edad)),
    'media': float(np.mean(edad)),
    'p75': float(np.percentile(edad, 75)),
    'max': float(np.max(edad)),
    'rango(ptp)': float(np.ptp(edad)),
    'var_muestral': float(np.var(edad, ddof=1)),    #degreace of fredom (grado de libertad)
    'std_muestral': float(np.std(edad, ddof=1)),    #variacióm estandar
    'cv': float(np.std(edad, ddof=1)/np.mean(edad)), #coeficiente de variación
}

summary_age
```

```python
# Frecuencias (NumPy)
freq_python = {'Si': int(np.count_nonzero(prog_py)), 'No': int(edad.size - np.count_nonzero(prog_py))}
freq_colab  = {'Si': int(np.count_nonzero(usa_colab)), 'No': int(edad.size - np.count_nonzero(usa_colab))}
freq_gh     = {'Si': int(np.count_nonzero(cuenta_gh)), 'No': int(edad.size - np.count_nonzero(cuenta_gh))}

print(f'Frecuencia de utilizar python: {freq_python}')
print(f'Frecuencia de utilizar colab: {freq_colab}')
print(f'Frecuencia de tener cuenta de github: {freq_gh}')
```

```python
# Tablas cruzadas (NumPy)
def tabla2x2(a, b):                       #función
    t = np.zeros((2,2), dtype=int)        #tabla bidi de enteros en cero
    t[1,1] = np.count_nonzero(a & b)      #asignara la cuenta de lo que no sea cero en un "and"
    t[1,0] = np.count_nonzero(a & ~b)     #multiplicacion and
    t[0,1] = np.count_nonzero(~a & b)
    t[0,0] = np.count_nonzero(~a & ~b)
    return t

tab_py_colab = tabla2x2(prog_py, usa_colab)   #comprueba si usa py y colab
tab_py_gh    = tabla2x2(prog_py, cuenta_gh)   #
tab_colab_gh = tabla2x2(usa_colab, cuenta_gh)

print(f'Tabla python vs colab:\n{tab_py_colab}')
print(f'Tabla python vs github:\n{tab_py_gh}')
print(f'Tabla colab vs github:\n{tab_colab_gh}')

```

```python
# Medias de edad por grupo (NumPy)
means_by_group = {
    'edad_promedio_Python': {'Si': float(np.mean(edad[prog_py])) if np.any(prog_py) else np.nan,
                             'No': float(np.mean(edad[~prog_py])) if np.any(~prog_py) else np.nan},
    'edad_promedio_Colab':  {'Si': float(np.mean(edad[usa_colab])) if np.any(usa_colab) else np.nan,
                             'No': float(np.mean(edad[~usa_colab])) if np.any(~usa_colab) else np.nan},
    'edad_promedio_GitHub': {'Si': float(np.mean(edad[cuenta_gh])) if np.any(cuenta_gh) else np.nan,
                             'No': float(np.mean(edad[~cuenta_gh])) if np.any(~cuenta_gh) else np.nan},
}

means_by_group
```

#### Mineria de datos

La Mineria de datos: es un proceso para analizar datos en volumenes, esto sirve para descubrir patrones, tendencias y correlaciones ocultas, asi de esta manera se transforma la información cruda en conocimiento util para predecir y tomar decisiones con estrategia en negocios o grandes empresas, ayudando al futuro de estas en cuanto a ventas o avances dentro de ellas.
